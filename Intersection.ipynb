```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
files = [".".join(f.split(".")[:-1]) for f in os.listdir("Responses")]
files.sort()
```

```python
dfs = [pd.read_csv("Responses/{}.csv".format(file)) for file in files]
```

```python
n = len(dfs)
arr = np.identity(n)

for i in range(n):
    for j in range(n):
        joint = pd.merge(dfs[i], dfs[j], how="inner", on=["Roll Number"])
        arr[i][j] = len(joint)/len(dfs[i])
```

```python
# Percentage overlap
sns.set(rc={"figure.figsize": (10, 7.5)})
sns.heatmap(arr*100, annot=True, linewidth=1, xticklabels=files, yticklabels=files)
```
